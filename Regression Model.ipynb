```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Algerin Fire.csv")
```

```python
df.info()
```

```python
x = df.drop('FWI',axis=1)
```

```python
y= df['FWI']
```

```python
y
```

```python
#Train test split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
x_test.shape,y_test.shape,x_train.shape,y_train.shape
```

```python
def corellation(dataset,threshold):
    col_corr = set()
    corr_martix = dataset.corr()
    for i in range (len(corr_martix.columns)):
        for j in range (i):
            if abs(corr_martix.iloc[i,j]) > threshold:
                colname = corr_martix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_feture = corellation(x_train,0.85)
```

```python
corr_feture
```

```python
x_train.drop(corr_feture,axis=1,inplace=True)
x_test.drop(corr_feture,axis=1,inplace=True)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_test_scaled
```

# Leanear Regresion model

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(x_train_scaled,y_train)
```

```python
y_pred = regression.predict(x_test_scaled)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y_test,y_pred)
mae
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,y_pred)
score
```
